## Sanity sanitizer

- Enforces alternating roles starting with human/user
- Drops trailing empty messages
- Trims any final unpaired human so the length is always even

```python
from typing import List, Dict, Tuple

HUMAN_ALIASES = {"human", "user"}
ASSISTANT_ALIASES = {"assistant", "gpt", "bot"}

def _norm_role(role: str) -> str | None:
    r = (role or "").strip().lower()
    if r in HUMAN_ALIASES:
        return "user"
    if r in ASSISTANT_ALIASES:
        return "assistant"
    return None

def _drop_trailing_empties(dialogs: List[Dict]) -> Tuple[List[Dict], int]:
    i = len(dialogs) - 1
    dropped = 0
    while i >= 0 and (dialogs[i].get("content") or "").strip() == "":
        dropped += 1
        i -= 1
    return dialogs[: i + 1], dropped

def sanitize_dialog_record(record: Dict, verbose: bool = True) -> Dict:
    rid = record.get("_id") or record.get("id") or "<no-id>"
    def emit(msg: str):
        if verbose:
            print(f"[sanitizer][{rid}] {msg}")
        logs.append(msg)

    logs: List[str] = []
    original = record.get("dialogs", [])
    before_len = len(original)

    # Normalize roles, preserve index for logging
    cleaned = []
    for idx, m in enumerate(original):
        role = _norm_role(m.get("role"))
        content = (m.get("content") or "")
        if role is None:
            emit(f"Dropped message #{idx} with unknown role '{m.get('role')}'.")
            continue
        cleaned.append({"role": role, "content": content.strip(), "_orig_idx": idx})

    # Drop trailing empties
    cleaned, n_empty1 = _drop_trailing_empties(cleaned)
    if n_empty1:
        emit(f"Dropped {n_empty1} trailing empty message(s).")

    # Ensure first is human
    drop_count_leading_assist = 0
    while cleaned and cleaned[0]["role"] != "user":
        m = cleaned.pop(0)
        drop_count_leading_assist += 1
        emit(f"Dropped leading assistant at orig #{m['_orig_idx']} to start with human.")
    if drop_count_leading_assist == before_len and before_len > 0:
        emit("All turns were assistants or invalid; dialog emptied.")

    # Enforce strict alternation
    alternating = []
    expected = "user"
    for j, m in enumerate(cleaned):
        if m["role"] == expected:
            alternating.append(m)
            expected = "assistant" if expected == "user" else "user"
        else:
            emit(
                f"Skipped out-of-order {m['role']} at cleaned pos #{j} "
                f"(orig #{m['_orig_idx']}), expected {expected}."
            )

    # Drop trailing empties again (could appear after alternation)
    alternating, n_empty2 = _drop_trailing_empties(alternating)
    if n_empty2:
        emit(f"Dropped {n_empty2} trailing empty message(s) after alternation.")

    # Ensure even length (trim final unpaired human)
    if len(alternating) % 2 == 1:
        last = alternating.pop()
        emit(f"Trimmed final unpaired {last['role']} (orig #{last.get('_orig_idx','?')}).")

    # Strip helper keys
    for m in alternating:
        m.pop("_orig_idx", None)

    # Validate
    def _is_valid(seq):
        if not seq:
            return False
        if seq[0]["role"] != "user":
            return False
        for i in range(1, len(seq)):
            if seq[i]["role"] == seq[i - 1]["role"]:
                return False
        return len(seq) % 2 == 0

    after_len = len(alternating)
    dropped = before_len - after_len
    if dropped > 0:
        emit(f"Summary: kept {after_len}/{before_len} turns (dropped {dropped}).")
    elif logs:
        emit(f"Summary: structure fixed with no net drop (still {after_len} turns).")

    record["dialogs"] = alternating
    record["is_valid_dialog"] = _is_valid(alternating)
    record["num_kept_turns"] = after_len
    record["sanitizer_logs"] = logs
    return record

# --- Hugging Face usage ---
# from datasets import load_dataset

data1 = data.map(lambda ex: sanitize_dialog_record(ex, verbose=True))
data1 = data1.filter(lambda ex: ex["is_valid_dialog"])
data1 = data1.remove_columns(["is_valid_dialog"])  # optional
data1
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer
from unsloth.chat_templates import get_chat_template

# 1. Load your dataset
#ds = RAW
ds = data

# 2. Load the base tokenizer for Gemma-3
tok = AutoTokenizer.from_pretrained("google/gemma-3-27b-it")

# 3. Wrap it with the gemma-3 chat template
tok = get_chat_template(
    tok,
    chat_template="gemma-3",            # select the gemma-3 template
)

# 4. Define a formatting fn that applies the template to each conversation
def format_for_gemma3(examples):
    convos = examples["dialogs"]  # list[dict(role,content)] per example
    texts = [
        tok.apply_chat_template(
            convo,
            tokenize=False,             # return raw string
            add_generation_prompt=False # no extra “assistant:” at end
        )
        for convo in convos
    ]
    return {"text": texts}

# 5. Map it over your dataset
ds = ds.map(format_for_gemma3, batched=True)

# Now ds["text"] holds strings like:
# <s>user: …<endofturn>model: …<endofturn>user: …<endofturn>model:
```

```python
from datasets import Dataset
from copy import deepcopy

# assume you already have:
# ds1_too_long: a Hugging Face Dataset with 'dialogs' and 'text'
# tokenizer: a HF tokenizer (e.g., AutoTokenizer.from_pretrained(...))

MAX_LEN = 2048

def build_text_from_dialogs(dialogs):
    """
    Robustly rebuild 'text' from a dialogs list.
    Handles common shapes: list[str] or list[dict] with role/content-ish keys.
    If it's already a string, just return it.
    """
    if isinstance(dialogs, str):
        return dialogs

    parts = []
    for turn in dialogs:
        if isinstance(turn, dict):
            role = turn.get("role") or turn.get("speaker") or turn.get("from") or "unknown"
            content = turn.get("content") or turn.get("value") or turn.get("text") or ""
            parts.append(f"<start_of_turn>{role}\n{content}\n<end_of_turn>")
        else:
            parts.append(str(turn))
    return "\n".join(parts)

def shrink_dialogs_until_fit(example, tokenizer, max_len=MAX_LEN):
    """
    Returns a modified example with trimmed dialogs/text/tokens,
    or None if the example must be dropped.
    """
    dialogs = example.get("dialogs", None)
    if not isinstance(dialogs, list):
        # can't trim meaningfully; drop if not a list or is missing
        return None

    # if already too short, drop
    if len(dialogs) <= 2:
        return None

    # work on a copy to avoid mutating the original
    dlg = list(dialogs)

    while True:
        # rebuild text from the current dialogs
        text = build_text_from_dialogs(dlg)

        # tokenize WITHOUT truncation so we can measure real length
        toks = tokenizer(text, add_special_tokens=True, truncation=False)

        if len(toks["input_ids"]) <= max_len:
            out = deepcopy(example)
            out["dialogs"] = dlg
            out["text"] = text
            out["tokens"] = toks
            return out

        # still too long → trim last two turns if possible
        if len(dlg) > 2:
            dlg = dlg[:-2]
        else:
            # cannot trim further → drop
            return None

def process_dataset(ds, tokenizer, max_len=MAX_LEN):
    processed = []
    for ex in ds:
        fixed = shrink_dialogs_until_fit(ex, tokenizer, max_len=max_len)
        if fixed is not None:
            processed.append(fixed)
    return Dataset.from_list(processed) if processed else Dataset.from_list([])

# usage:
ds1_too_long_fixed = process_dataset(ds1_too_long_filtered, tokenizer, max_len=2048)
ds1_too_long_fixed
```

```python
from datasets import load_dataset
from transformers import AutoTokenizer
from unsloth.chat_templates import get_chat_template


# 2. Load the base tokenizer for Gemma-3
tok = AutoTokenizer.from_pretrained("google/gemma-3-27b-it")

# 3. Wrap it with the gemma-3 chat template
tok = get_chat_template(
    tok,
    chat_template="gemma-3",            # select the gemma-3 template
)

# 4. Define a formatting fn that applies the template to each conversation
def format_for_gemma3(examples):
    convos = examples["dialogs"]  # list[dict(role,content)] per example
    texts = [
        tok.apply_chat_template(
            convo,
            tokenize=False,             # return raw string
            add_generation_prompt=False # no extra “assistant:” at end
        )
        for convo in convos
    ]
    return {"text": texts}

# 5. Map it over your dataset
ds = ds.map(format_for_gemma3, batched=True)

# Now ds["text"] holds strings like:
# <s>user: …<endofturn>model: …<endofturn>user: …<endofturn>model:
```

```python
# first we keep only the ones that have more than one turn, otherwise we wont be able to further trim them
ds1_too_long_filtered = ds1_too_long.filter(lambda example: len(example['dialogs'])>2)
ds1_too_long_filtered
```

```python
from datasets import Dataset
from copy import deepcopy

# assume you already have:
# ds1_too_long: a Hugging Face Dataset with 'dialogs' and 'text'
# tokenizer: a HF tokenizer (e.g., AutoTokenizer.from_pretrained(...))

MAX_LEN = 2048

def build_text_from_dialogs(dialogs):
    """
    Robustly rebuild 'text' from a dialogs list.
    Handles common shapes: list[str] or list[dict] with role/content-ish keys.
    If it's already a string, just return it.
    """
    if isinstance(dialogs, str):
        return dialogs

    parts = []
    for turn in dialogs:
        if isinstance(turn, dict):
            role = turn.get("role") or turn.get("speaker") or turn.get("from") or "unknown"
            content = turn.get("content") or turn.get("value") or turn.get("text") or ""
            parts.append(f"<start_of_turn>{role}\n{content}\n<end_of_turn>")
        else:
            parts.append(str(turn))
    return "\n".join(parts)

def shrink_dialogs_until_fit(example, tokenizer, max_len=MAX_LEN):
    """
    Returns a modified example with trimmed dialogs/text/tokens,
    or None if the example must be dropped.
    """
    dialogs = example.get("dialogs", None)
    if not isinstance(dialogs, list):
        # can't trim meaningfully; drop if not a list or is missing
        return None

    # if already too short, drop
    if len(dialogs) <= 2:
        return None

    # work on a copy to avoid mutating the original
    dlg = list(dialogs)

    while True:
        # rebuild text from the current dialogs
        text = build_text_from_dialogs(dlg)

        # tokenize WITHOUT truncation so we can measure real length
        toks = tokenizer(text, add_special_tokens=True, truncation=False)

        if len(toks["input_ids"]) <= max_len:
            out = deepcopy(example)
            out["dialogs"] = dlg
            out["text"] = text
            out["tokens"] = toks
            return out

        # still too long → trim last two turns if possible
        if len(dlg) > 2:
            dlg = dlg[:-2]
        else:
            # cannot trim further → drop
            return None

def process_dataset(ds, tokenizer, max_len=MAX_LEN):
    processed = []
    for ex in ds:
        fixed = shrink_dialogs_until_fit(ex, tokenizer, max_len=max_len)
        if fixed is not None:
            processed.append(fixed)
    return Dataset.from_list(processed) if processed else Dataset.from_list([])

# usage:
ds1_too_long_fixed = process_dataset(ds1_too_long_filtered, tokenizer, max_len=2048)
ds1_too_long_fixed
```
